````python
"""**Task Description:**
[Provide a detailed description of the math word problem here. Include all necessary information such as numbers, relationships, constraints, and the specific question that needs to be answered.]

**Requirements:**
1. Read and interpret the provided math word problem.
2. Translate the problem into a Python function that takes relevant inputs and returns the solution.
3. Always write your Python code in between ```python and ```.
4. Always have a print statement in the python code to print the response required from the python program.
5. Include comments in the code to explain the logic and steps taken to solve the problem.
6. Provide a brief explanation of how the code works and how it solves the given problem.
7. Once a code step is reached, stop to get the output of the code (an external python executor will execute and 
   provide the output).
8. Once you are confident about the answer, provide our final answer in last line as Final_answer: (your final answer). 
"""
````

````python
import openai
import traceback
import re
import time
import sys
from io import StringIO
import func_timeout


class Code_interpreter:
    
    def __init__(self, deployment_name="gpt-4-turbo"):
                
        openai.api_type = ""
        openai.api_base = ''
        openai.api_version = ''
        openai.api_key = ''
        
        self.deployment_name = deployment_name
        self.reset()
        
        
    
    def reset(self):
        self.cost = {
                      "completion_tokens": 0,
                      "prompt_tokens": 0,
                      "total_tokens": 0
                    }
        
        self.messages = [ {"role": "system", "content": """**Task Description:**
You are given a math word problem that involves several steps to reach the solution. The problem includes numbers, relationships, and constraints. You must read and understand the problem, then create a Python function that takes the necessary inputs and returns the correct solution. The problem may involve operations such as addition, subtraction, multiplication, division, or more complex mathematical concepts.

**Requirements:**
1. Read and interpret the provided math word problem.
2. Break down the problem into smaller parts if necessary and solve each part step by step.
3. Translate each part of the problem into a Python snippets or a series of Python functions that take relevant inputs and return the solutions.
4. Write your Python code in snippets between ```python and ``` as needed, and wait for external executor to get intermediate results for each snippet.
5. Include Print statement in each Python program snippet to get output from external executor.
6. Include comments in the code to explain the logic and steps taken to solve each part of the problem.
7. Provide a brief explanation of how each code snippet works and how it contributes to solving the given problem.
8. **Once a code step is reached, stop to get the output of the code from an external python executor and proceed from there**.
9. After solving all parts of the problem and reaching the final answer, provide the final answer in the last line as Final_answer: (your final answer).

Note: ** You cannot skip step 3 and 4. **

Please provide the math word problem to proceed with the solution.
"""}]
        
    def request_openai(self, temp, count=0):
        
        try:
            res = openai.ChatCompletion.create(
                engine=self.deployment_name,
                messages=self.messages,
                temperature=temp
            )
            return res
        except Exception as e:
            print(f"Exception {e}")
            if count>5:
                print(f"Exception e tried 5 times.")
            time.sleep(5)
            return self.request_openai(temp, count+1)
        
    def generate_response(self, message: str = "", temp : float = 0.0):
        
        if message != "":
            self.messages.append(
            {"role": "user", "content": message}
            )
        
        response = self.request_openai(temp, 0)
        
        self.messages.append({
        "role": "assistant", "content": response['choices'][0]['message']['content']
        })
        
        ## Tracking token usage..
        self.cost['completion_tokens'] += response['usage']['completion_tokens']
        self.cost['prompt_tokens'] += response['usage']['prompt_tokens']
        self.cost['total_tokens'] += response['usage']['total_tokens']
        
        
        
        return response['choices'][0]['message']['content']
    
    def extract_code_and_execute(self, response):
        
        try:
            code = re.findall(r"```python[^`]*```", response)[-1].replace("```python", "").replace("```", "")
            if code[0] == "\n":
                code = code[1:]
                
            execution_output = self.safe_execute(code)
            self.messages.append({
                                "role":  "user",
                                "content": "Following is the output of the execution for the given code:\n\n" + execution_output
                                })
        except:
            self.messages.append({
                                "role":  "user",
                                "content": "Couldn't find the python code to execute (OR) syntax of code between ```python and ``` is not followed."
                                })

        
        
    def safe_execute(self, code_string: str, keys=None):
        def execute(x):
            try:
                old_stdout = sys.stdout
                redirected_output = sys.stdout = StringIO()
                exec(x)
                sys.stdout = old_stdout
                return redirected_output.getvalue()
                # locals_ = locals()
                # if keys is None:
                #     return locals_.get('ans', None)
                # else:
                #     return [locals_.get(k, None) for k in keys]
            except Exception:
                sys.stdout = old_stdout
                return traceback.format_exc()
        try:
            ans = func_timeout.func_timeout(10, execute, args=(code_string,))
        except func_timeout.FunctionTimedOut:
            ans = "Execution timeout.."

        return ans
    
    def solve(self, question):
        
        self.reset()
        
        ## Step 1. Generation.
        res = self.generate_response(question)
        print("generation completed")
        count = 0
        while (not "Final_answer" in res) and count<5:
            
            self.extract_code_and_execute(res)
            print("Execution completed..")
            res = self.generate_response()
            print("generation completed")
        
            count += 1
        
        ## Step 2. Verify the solution.
        print("\n Verification Started..")
        count = 0
        verify = """Let's verify the solution using the python code. Finally give Verification_Label: (True, False, uncertain) by comparing the solution. 
        
        If you are done with verification in the last line say "TERMINATE"
        """
        res = self.generate_response(verify)
        while (not "TERMINATE" in res) and count<5:
            
            self.extract_code_and_execute(res)
            print("Execution completed..")
            res = self.generate_response()
            print("generation completed")
        
            count += 1
        
    
    def pretty_print(self):
        
        for i in self.messages:
            print(f"{i['role']}\n\n{i['content']}")
            print("\n" + "-"*60 + "\n")
````

````python
c = Code_interpreter()
````

````python
c.solve("Find all positive integer values of $c$ such that the equation $x^2-7x+c=0$ only has roots that are real and rational. Express them in decreasing order, separated by commas.")
````
