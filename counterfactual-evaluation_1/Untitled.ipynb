```python
import json

def get_cost(items):
    
    count = {'prompt_tokens': 0,
      'completion_tokens': 0,
      'total_tokens': 0}

    for i in range(len(df)):
        usage = json.loads(df.iloc[i]['usage'])
        count['prompt_tokens'] = count['prompt_tokens'] + usage['prompt_tokens']
        count['completion_tokens'] = count['completion_tokens'] + usage['completion_tokens'] 
        count['total_tokens'] = count['total_tokens'] + usage['total_tokens'] 

    dollars = 0.06 * (count['prompt_tokens']) / 1000  +  0.12 * (count['completion_tokens']) / 1000
    rupees = dollars * 83.3
    print(f"Total Cost: ${dollars} or Rs.{rupees}")
    
def get_average(l):
    return sum(l)/len(l)
```
