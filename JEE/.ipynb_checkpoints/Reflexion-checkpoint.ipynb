```python
import dataclasses
from typing import Union, List, Optional, Callable, NamedTuple, Tuple, Literal

from tenacity import (
    retry,
    stop_after_attempt,  # type: ignore
    wait_random_exponential,  # type: ignore
)
import openai

openai.api_type = "azure"
openai.api_base = ""
openai.api_version = ""
openai.api_key = ""

MessageRole = Literal["system", "user", "assistant"]

@dataclasses.dataclass()
class Message():
    role: MessageRole
    content: str

@retry(wait=wait_random_exponential(min=1, max=180), stop=stop_after_attempt(6))
def gpt_chat(
    model: str,
    messages: List[Message],
    max_tokens: int = 1024,
    temperature: float = 0.7,
    num_comps=1,
    logger = None
) -> Union[List[str], str]:
    
    p_string = """Input to GPT4:\n"""
    for message in messages:
        p_string += f"""\n{message.role}\n{message.content}\n"""
    logger.info(p_string)
    response = openai.ChatCompletion.create(
        engine=model,
        messages=[dataclasses.asdict(message) for message in messages],
        max_tokens=max_tokens,
        temperature=temperature,
        top_p=1,
        frequency_penalty=0.0,
        presence_penalty=0.0,
        n=num_comps
    )
    
    if num_comps == 1:   # no.of completions ==  1.
        logger.info(f"\nresponse: \n {response.choices[0].message.content}")
        return response.choices[0].message.content  # type: ignore

    logger.info(f"\nresponse: \n {[choice.message.content for choice in response.choices]}")
    return [choice.message.content for choice in response.choices]  # type: ignore


class GPT4():
    def __init__(self, name: str = "gpt-4-32k", temperature: float = 0.7, max_tokens: float = 4000,  logger = None):
        self.name = name
        self.logger = logger
        self.temperature = temperature
        self.max_tokens = max_tokens

    def __repr__(self) -> str:
        return f'{self.name}'

    def generate_chat(self, messages: List[Message], max_tokens: int = 1024, temperature: float = 0.2, num_comps: int = 1) -> Union[List[str], str]:
        return gpt_chat(self.name, messages, self.max_tokens, self.temperature, num_comps, logger = self.logger)
```

```python
from typing import Union, List, Optional, Callable, NamedTuple, Tuple
from model import Message

#############################  Reflexion prompts   #########################

PY_SELF_REFLECTION_CHAT_INSTRUCTION = "You are a experienced IIT Professor. You will be given an answer you provided to a question along with the feedback. Your goal is to write a few sentences to explain why your answer is wrong as indicated by the feedback. You will need this as a hint when you try again later. Only provide the few sentences description why the above answer is wrong. Do not provide updated answer."

PY_SELF_REFLECTION_FEW_SHOT = """
Example 1:
[Question]:
Let $f: \mathbb{R} \rightarrow \mathbb{R}$ be a differentiable function such that $f(0)=0, f\left(\frac{\pi}{2}\right)=3$ and $f^{\prime}(0)=1$. If

\[
g(x)=\int_{x}^{\frac{\pi}{2}}\left[f^{\prime}(t) \operatorname{cosec} t-\cot t \operatorname{cosec} t f(t)\right] d t
\]

for $x \in\left(0, \frac{\pi}{2}\right]$, then what is the $\lim _{x \rightarrow 0} g(x)$?

[Answer Generated]:
Understand the problem and given information

We are given a differentiable function $f: \mathbb{R} \rightarrow \mathbb{R}$ with $f(0)=0, f\left(\frac{\pi}{2}\right)=3$ and $f^{\prime}(0)=1$. We are also given a function $g(x)$ defined as:

\[
g(x)=\int_{x}^{\frac{\pi}{2}}\left[f^{\prime}(t) \operatorname{cosec} t-\cot t \operatorname{cosec} t f(t)\right] d t
\]

for $x \in\left(0, \frac{\pi}{2}\right]$. Our goal is to find the limit of $g(x)$ as $x$ approaches $0$.

Step 2: Apply the Fundamental Theorem of Calculus

The Fundamental Theorem of Calculus states that if $F(x)$ is an antiderivative of $f(x)$ on an interval $[a, b]$, then

\[
\int_{a}^{b} f(x) dx = F(b) - F(a)
\]

Applying this theorem to the given integral, we get:

\[
g(x) = G\left(\frac{\pi}{2}\right) - G(x)
\]

where $G(t)$ is an antiderivative of $f^{\prime}(t) \operatorname{cosec} t-\cot t \operatorname{cosec} t f(t)$.

Step 3: Differentiate $g(x)$ with respect to $x$

To find the limit of $g(x)$ as $x$ approaches $0$, we can differentiate $g(x)$ with respect to $x$ and then evaluate the derivative at $x=0$. Using the chain rule, we get:

\[
g^{\prime}(x) = -G^{\prime}(x) = -\left[f^{\prime}(x) \operatorname{cosec} x-\cot x \operatorname{cosec} x f(x)\right]
\]

Step 4: Evaluate the derivative at $x=0$

Now, we need to evaluate $g^{\prime}(0)$. We know that $f(0)=0$ and $f^{\prime}(0)=1$. Also, $\cot(0)$ and $\operatorname{cosec}(0)$ are undefined. However, we can use L'Hôpital's rule to evaluate the limit:

\[
\lim_{x \rightarrow 0} \frac{\cot x \operatorname{cosec} x f(x)}{x} = \lim_{x \rightarrow 0} \frac{-\csc^2 x f(x) + \cot x \operatorname{cosec} x f^{\prime}(x)}{1}
\]

Since $f(0)=0$ and $f^{\prime}(0)=1$, the limit evaluates to:

\[
\lim_{x \rightarrow 0} \frac{-\csc^2 x f(x) + \cot x \operatorname{cosec} x f^{\prime}(x)}{1} = 0
\]

Thus, $g^{\prime}(0) = -\left[f^{\prime}(0) \operatorname{cosec} 0 - 0\right] = -1$.

Step 5: Integrate the derivative and find the limit

Now that we have the derivative of $g(x)$, we can integrate it to find $g(x)$:

\[
g(x) = -x + C
\]

We know that $g\left(\frac{\pi}{2}\right) = 0$, so we can solve for $C$:

\[
0 = -\frac{\pi}{2} + C \Rightarrow C = \frac{\pi}{2}
\]

Thus, $g(x) = -x + \frac{\pi}{2}$. Finally, we can find the limit of $g(x)$ as $x$ approaches $0$:

\[
\lim_{x \rightarrow 0} g(x) = -0 + \frac{\pi}{2} = \frac{\pi}{2}
\]

Final_answer: 2

[Checklist feedback]:
Satisfied points:
    1. The student has correctly identified the given information: f(0)=0, f(π/2)=3, and f'(0)=1.
    2. The student has correctly set up the integral for the function g(x): g(x)=∫[x,π/2][f'(t) cosec t-cot t cosec t f(t)] dt
Failed Points:
     3. The student has applied integration by parts to solve the integral, using u=f(t) and dv=[f'(t) cosec t-cot t cosec t f(t)] dt.
    4. The student has correctly calculated the integral of dv to find v and the derivative of u to find du.
    5. The student has applied the integration by parts formula: g(x) = uv|_x^(π/2) - ∫[x,π/2] u dv.
    6. The student has correctly evaluated the limits of integration and simplified the expression for g(x).
    7. The student has correctly calculated the limit lim _{x → 0} g(x) based on their simplified expression for g(x).
    8. The student has provided the correct numerical value for the limit as their final answer.


[Self-reflexion] :
My answer is wrong because I did not correctly apply integration by parts to solve the integral. I mistakenly used the Fundamental Theorem of Calculus and then tried to differentiate the function g(x) to find its limit as x approaches 0. In doing so, I ended up with an incorrect expression for g(x) and consequently an incorrect final answer. I should have used integration by parts with u = f(t) and dv = [f'(t) cosec t - cot t cosec t f(t)] dt, and then followed the subsequent steps to correctly calculate the integral of dv, find v and du, and apply the integration by parts formula.

END OF EXAMPLES.
"""


#############################   Answer Generation #########################

PY_SIMPLE_COT_INSTRUCTION = "You are an Experienced Professor from IIT Madras. You will Answer the given question step by step and finally give the answer."

# generation using reflexion.

PY_REFLEXION_CHAT_INSTRUCTION = "You are an IIT Student. You will be given your past solution, a series of check list which are either passed or failed, and a hint to change the solution appropriately. Write your full implementation freshly using these feedbacks."

PY_REFLEXION_FEW_SHOT_ADD = '''Example 1:
Solution:
Step 1: Understand the problem and given information

We are given a differentiable function $f: \mathbb{R} \rightarrow \mathbb{R}$ with $f(0)=0, f\left(\frac{\pi}{2}\right)=3$ and $f^{\prime}(0)=1$. We are also given a function $g(x)$ defined as:

\[
g(x)=\int_{x}^{\frac{\pi}{2}}\left[f^{\prime}(t) \operatorname{cosec} t-\cot t \operatorname{cosec} t f(t)\right] d t
\]

Our goal is to find the limit of $g(x)$ as $x$ approaches $0$.

Step 2: Apply the Fundamental Theorem of Calculus

The Fundamental Theorem of Calculus states that if $F(x)$ is an antiderivative of $f(x)$ on an interval $[a, b]$, then:

\[
\int_{a}^{b} f(x) dx = F(b) - F(a)
\]

Applying this theorem to the given function $g(x)$, we get:

\[
g(x) = F\left(\frac{\pi}{2}\right) - F(x)
\]

where $F(t)$ is an antiderivative of $f^{\prime}(t) \operatorname{cosec} t-\cot t \operatorname{cosec} t f(t)$.

Step 3: Find the antiderivative $F(t)$

To find the antiderivative $F(t)$, we can use integration by parts. Let $u = f(t)$ and $dv = -\cot t \operatorname{cosec} t dt$. Then, $du = f^{\prime}(t) dt$ and $v = \operatorname{cosec} t$. Now, we can rewrite the integral as:

\[
F(t) = \int \left[f^{\prime}(t) \operatorname{cosec} t-\cot t \operatorname{cosec} t f(t)\right] dt = \int u dv
\]

Using integration by parts, we get:

\[
F(t) = uv - \int v du = -f(t) \operatorname{cosec} t - \int (-\operatorname{cosec} t) f^{\prime}(t) dt
\

Step 4: Evaluate the limit

Now, we can find the limit of $g(x)$ as $x$ approaches $0$:

\[
\lim_{x \rightarrow 0} g(x) = \lim_{x \rightarrow 0} \left[F\left(\frac{\pi}{2}\right) - F(x)\right]
\]

Using the given information, we know that $f(0) = 0$ and $f^{\prime}(0) = 1$. Thus, as $x$ approaches $0$, $F(x)$ approaches $0$ and $F^{\prime}(x)$ approaches $1$. Therefore, we have:

\[
\lim_{x \rightarrow 0} g(x) = F\left(\frac{\pi}{2}\right) - 0 = F\left(\frac{\pi}{2}\right)
\]

Since $f\left(\frac{\pi}{2}\right) = 3$, we can evaluate $F\left(\frac{\pi}{2}\right)$ as:

\[
F\left(\frac{\pi}{2}\right) = -3 \operatorname{cosec} \left(\frac{\pi}{2}\right) - \int_{0}^{\frac{\pi}{2}} (-\operatorname{cosec} t) dt = -3(1) - \int_{0}^{\frac{\pi}{2}} (-\operatorname{cosec} t) dt
\]

Now, we need to evaluate the integral $\int_{0}^{\frac{\pi}{2}} (-\operatorname{cosec} t) dt$. This integral can be evaluated using a substitution. Let $u = \cos t$, then $du = -\sin t dt$. The integral becomes:

\[
\int_{0}^{\frac{\pi}{2}} (-\operatorname{cosec} t) dt = \int_{1}^{0} \frac{1}{u} du = \left[-\ln |u|\right]_{1}^{0} = -\ln 1 + \ln 0 = 0
\]

Therefore, the limit of $g(x)$ as $x$ approaches $0$ is:

\[
\lim_{x \rightarrow 0} g(x) = F\left(\frac

feedback:
Satisfied Points:
1. The student has correctly identified the given information: f(0)=0, f(\\frac{\\pi}{2})=3, and f^\\prime(0)=1.
2. The student has understood the definition of the function g(x) and its relation to the function f(x).
3. The student has applied the Fundamental Theorem of Calculus to find the derivative of g(x) with respect to x.

Failed Points:
 4. The student has correctly simplified the expression for g'(x) using the given information and trigonometric identities.
5. The student has evaluated the limit \\lim _{x \\rightarrow 0} g'(x) using the appropriate limit rules and trigonometric limits.
6. The student has used the relationship between g'(x) and g(x) to find the limit \\lim _{x \\rightarrow 0} g(x).
7. The student has provided a final answer for the limit, which is consistent with their previous calculations and the given problem constraints.

[reflection on previous solution]:
My answer is wrong because I did not correctly apply integration by parts to solve the integral. I mistakenly used the Fundamental Theorem of Calculus and then tried to differentiate the function g(x) to find its limit as x approaches 0. In doing so, I ended up with an incorrect expression for g(x) and consequently an incorrect final answer. I should have used integration by parts with u = f(t) and dv = [f'(t) cosec t - cot t cosec t f(t)] dt, and then followed the subsequent steps to correctly calculate the integral of dv, find v and du, and apply the integration by parts formula.

Problem:
Let $f: \mathbb{R} \rightarrow \mathbb{R}$ be a differentiable function such that $f(0)=0, f\left(\frac{\pi}{2}\right)=3$ and $f^{\prime}(0)=1$. If

\[
g(x)=\int_{x}^{\frac{\pi}{2}}\left[f^{\prime}(t) \operatorname{cosec} t-\cot t \operatorname{cosec} t f(t)\right] d t
\]

for $x \in\left(0, \frac{\pi}{2}\right]$, then what is the $\lim _{x \rightarrow 0} g(x)$?

[improved solution]:

Step 1: Understand the problem and given information

We are given a differentiable function $f: \mathbb{R} \rightarrow \mathbb{R}$ with $f(0)=0, f\left(\frac{\pi}{2}\right)=3$ and $f^{\prime}(0)=1$. We are also given a function $g(x)$ defined as:

\[
g(x)=\int_{x}^{\frac{\pi}{2}}\left[f^{\prime}(t) \operatorname{cosec} t-\cot t \operatorname{cosec} t f(t)\right] d t
\]

Our goal is to find the limit of $g(x)$ as $x$ approaches $0$.

Step 2: Apply integration by parts

To find the integral, we can use integration by parts. Let $u = f(t)$ and $dv = [f'(t) \operatorname{cosec} t - \cot t \operatorname{cosec} t f(t)] dt$. Then, $du = f^{\prime}(t) dt$ and $v = \int [f'(t) \operatorname{cosec} t - \cot t \operatorname{cosec} t f(t)] dt$. Now, we can rewrite the integral as:

\[
g(x) = \int_{x}^{\frac{\pi}{2}}\left[f^{\prime}(t) \operatorname{cosec} t-\cot t \operatorname{cosec} t f(t)\right] d t = \int_{x}^{\frac{\pi}{2}} u dv
\]

Using integration by parts, we get:

\[
g(x) = uv\Big|_{x}^{\frac{\pi}{2}} - \int_{x}^{\frac{\pi}{2}} v du = -f(t) \operatorname{cosec} t\Big|_{x}^{\frac{\pi}{2}} - \int_{x}^{\frac{\pi}{2}} (-\operatorname{cosec} t) f^{\prime}(t) dt
\]

Step 3: Simplify the expression for $g(x)$

Now, we can simplify the expression for $g(x)$ using the given information and trigonometric identities:

\[
g(x) = -f\left(\frac{\pi}{2}\right) \operatorname{cosec} \left(\frac{\pi}{2}\right) + f(x) \operatorname{cosec}(x) + \int_{x}^{\frac{\pi}{2}} \operatorname{cosec}(t) f^{\prime}(t) dt
\]

Since $f\left(\frac{\pi}{2}\right) = 3$, we have:

\[
g(x) = -3(1) + f(x) \operatorname{cosec}(x) + \int_{x}^{\frac{\pi}{2}} \operatorname{cosec}(t) f^{\prime}(t) dt = -3 + f(x) \operatorname{cosec}(x) + \int_{x}^{\frac{\pi}{2}} \operatorname{cosec}(t) f^{\prime}(t) dt
\]

Step 4: Evaluate the limit

Now, we can find the limit of $g(x)$ as $x$ approaches $0$:

\[
\lim_{x \rightarrow 0} g(x) = \lim_{x \rightarrow 0} \left[-3 + f(x) \operatorname{cosec}(x) + \int_{x}^{\frac{\pi}{2}} \operatorname{cosec}(t) f^{\prime}(t) dt\right]
\]

Using the given information, we know that $f(0) = 0$ and $f^{\prime}(0) = 1$. Thus, as $x$ approaches $0$, the first two terms in the expression for $g(x)$ approach $-3$. The limit of the integral term can be evaluated using the Leibniz rule for differentiating under the integral sign:

\[
\frac{d}{dx} \int_{x}^{\frac{\pi}{2}} \operatorname{cosec}(t) f^{\prime}(t) dt = -\operatorname{cosec}(x) f^{\prime}(x)
\]

Since $f^{\prime}(0) = 1$, we have:

\[
\lim_{x \rightarrow 0} \frac{d}{dx} \int_{x}^{\frac{\pi}{2}} \operatorname{cosec}(t) f^{\prime}(t) dt = -\operatorname{cosec}(0) \cdot 1 = 0
\]

Therefore, the limit of the integral term as $x$ approaches $0$ is also $0$. Combining these results, we find the limit of $g(x)$ as $x$ approaches $0$:

\[
\lim_{x \rightarrow 0} g(x) = -3 + 0 + 0 = -3
\]

So, the limit of $g(x)$ as $x$ approaches $0$ is $-3$.

END OF EXAMPLES.
'''

#############################  Generate check list  ########################
PY_CHECKLIST_INSTRUCTION = "You are an Experienced Professor from IIT Madras. You will be given a Question from JEE paper. Generate a list of valid points which are Mutually Exclusive and Collectively Exhaustive, Which will be used to check whether an answer is correct or not."

PY_CHECKLIST_FEWSHOT = """Examples:
Problem:
In a historical experiment to determine Planck's constant, a metal surface was irradiated with light of different wavelengths. The emitted photoelectron energies were measured by applying a stopping potential. The relevant data for the wavelength $(\lambda)$ of incident light and the corresponding stopping potential $\left(V_{0}\right)$ are given below :

\begin{center}

\begin{tabular}{cc}

\hline

$\lambda(\mu \mathrm{m})$ & $V_{0}($ Volt $)$ \\

\hline

0.3 & 2.0 \\

0.4 & 1.0 \\

0.5 & 0.4 \\

\hline

\end{tabular}

\end{center}

Given that $c=3 \times 10^{8} \mathrm{~m} \mathrm{~s}^{-1}$ and $e=1.6 \times 10^{-19} \mathrm{C}$, Planck's constant (in units of $J \mathrm{~s}$ ) found from such an experiment is

(A) $6.0 \times 10^{-34}$

(B) $6.4 \times 10^{-34}$

(C) $6.6 \times 10^{-34}$

(D) $6.8 \times 10^{-34}$


Checklists:
To determine if an answer is correct for this problem, the following mutually exclusive and collectively exhaustive points must be checked:

1. The student has used the appropriate formula to determine Planck's constant: $h = \frac{eV_{0}\lambda}{c(1 - \frac{\lambda_{1}}{\lambda})}$
2. The student has correctly converted the given wavelengths from micrometers to meters: $\lambda = 10^{-6} \times \lambda(\mu m)$
3. The student has correctly calculated the difference in wavelengths for each data point: $\Delta \lambda = \lambda - \lambda_{1}$, where $\lambda_{1}$ is the smallest given wavelength (0.3 micrometers).
4. The student has correctly calculated the values of $h$ for each data point using the formula and the given values of $c$ and $e$.
5. The student has either averaged the calculated values of $h$ or determined that the values are consistent enough to select one of them as the final answer.
6. The student has selected the correct option from the given choices (A, B, C, or D) based on their calculated value(s) of Planck's constant.

END OF EXAMPLES
"""

#####################################  checking whether generated answer satisfies all points  #################

PY_SATISFIES_INSTRUCTION = """
You are a Professor in IIT Madras with 20 years of experience. You will be given a question, an answer from a student and a checklist to verify whether the answer follows all the checklist properly. Your have to explain which points are satisfied and which are violated. 
Give the response in strictly JSON format which can be parsed by Python json library, with one key as "explanation" and the  explanation, the second key as "SatisfiedPoints" and the Points which are satisfied (Complete points not just numbers), the third key as "FailedPoints" the points which are not satisfied (Complete points not just numbers)  as value and "FinalLabel" finally give "YES" or "NO" whether all points are satisfied or not in a new line.
"""




class PyGenerator:
    def self_reflexion(self, q: str, cur_answer: str, feedback:str, model) -> str:
        # reflection generation
        
        return gen_reflexion(
                            q,
                            cur_answer,
                            feedback,
                            model,
                            instruction=PY_SELF_REFLECTION_CHAT_INSTRUCTION,
                            few_shot_examples=PY_SELF_REFLECTION_FEW_SHOT
                            )
        

    def generate_answer(
        self,
        Question: str,
        model,
        strategy: str = 'simple',
        prev_answer: Optional[str] = None,
        feedback: Optional[str] = None,
        self_reflection: Optional[str] = None,
        num_comps: int = 1,
        temperature: float = 0.0,
    ) -> Union[str, List[str]]:
        return generic_generate_func_impl(
                Question=Question,
                model=model,
                strategy=strategy,
                prev_answer=prev_answer,
                feedback=feedback,
                self_reflection=self_reflection,
                num_comps=num_comps,
                temperature=temperature,
                reflexion_chat_instruction= PY_REFLEXION_CHAT_INSTRUCTION,
                reflexion_few_shot= PY_REFLEXION_FEW_SHOT_ADD,
                simple_chat_instruction= PY_SIMPLE_COT_INSTRUCTION
            )
    
    def Evaluate(self, Question: str, Answer: str, checklist: str, model):
        return check_if_all_satisfied(Question, Answer, checklist, model, PY_SATISFIES_INSTRUCTION)

    def generate_checklist(self, Question: str, model, max_num_tests: int = 5) -> List[str]:
        return generic_generate_internal_tests(
            Question=Question,
            model=model,
            max_num_tests=max_num_tests,
            test_generation_few_shot=PY_CHECKLIST_FEWSHOT,
            test_generation_chat_instruction=PY_CHECKLIST_INSTRUCTION
        )

def generic_generate_func_impl(
    Question: str,
    model,
    strategy: str,
    prev_answer,
    feedback,
    self_reflection,
    num_comps,
    temperature,
    reflexion_chat_instruction: str,
    reflexion_few_shot: str,
    simple_chat_instruction: str = PY_SIMPLE_COT_INSTRUCTION
) -> Union[str, List[str]]:
    
    if strategy != "reflexion" and strategy != "simple":
        raise ValueError(
            f"Invalid strategy: given `{strategy}` but expected one of `reflexion` or `simple`")
    if strategy == "reflexion" and (prev_answer is None or feedback is None or self_reflection is None):
        raise ValueError(
            f"Invalid arguments: given `strategy=reflexion` but `prev_func_impl`, `feedback`, or `self_reflection` is None")
    
    if strategy == 'simple':
        system_prompt = f"{simple_chat_instruction}"
        messages = [
            Message(
                role="system",
                content=system_prompt,
            ),
            Message(
                role="user",
                content=Question,
            ),
        ]
        answer = model.generate_chat(messages=messages, num_comps=num_comps, temperature=temperature)
    else:  # reflexion.
        system_prompt = f"{reflexion_chat_instruction}"
        messages = [
            Message(
                role="system",
                content=system_prompt,
            ),
            Message(
                role="user", # TODO: check this
                content=reflexion_few_shot,
            ),
            Message(
                role="assistant",
                content=f"""Solution:\n{prev_answer}""",
            ),
            Message(
                role="user",
                content=f"feedback:\n{feedback}\n\n[reflection on previous solution]:",
            ),
            Message(
                role="assistant",
                content=self_reflection,
            ),
            Message(
                role="user",
                content=f"[improved solution]:\nProblem: {Question}",
            ),
        ]
        answer = model.generate_chat(messages=messages, num_comps=num_comps, temperature=temperature)
        

    if num_comps == 1:
        assert isinstance(answer, str)
        return answer

    else:
        return answer
    
    

def generic_generate_internal_tests(
        Question: str,
        model,
        max_num_tests: int,
        test_generation_few_shot: str,
        test_generation_chat_instruction: str
) -> List[str]:
    """Generates tests for a function."""
    
    messages = [
        Message(
            role="system",
            content=test_generation_chat_instruction,
        ),
        Message(
            role="user",
            content=f"{test_generation_few_shot}\n\nProblem:\n{Question}\n\nChecklists:",
        )
    ]
    
    output = model.generate_chat(messages=messages, max_tokens=1024)
    
    return output
    
    
def check_if_all_satisfied(Question: str,
                           previous_answer: str,
                           checklist: str, 
                           model,
                           instruction: str = PY_SATISFIES_INSTRUCTION,
                          ):
    
    messages = [
        Message(
            role="system",
            content=instruction,
        ),
        Message(
            role="user",
            content=f"Problem:\n{Question}",
        ),
        Message(
            role="assistant",
            content=f"Solution Provided:\n{previous_answer}",
        ),
        Message(
            role="user",
            content=f"Checklists:\n{checklist}\n\n",
        )
    ]
    
    output = model.generate_chat(messages=messages, max_tokens=2024)
    
    return output


def gen_reflexion(
                q: str,
                cur_answer:str,
                feedback:str,
                model:str,
                instruction:str,
                few_shot_examples:str,
                ):

    if few_shot_examples is not None:
        messages = [
            Message(
                role="system",
                content=instruction,
            ),
            Message(
                role="user",
                content=f'{few_shot_examples}\n\n[Question]:\n{q}\n\n[Answer Generated]:\n{cur_answer}\n\n[Checklist feedback]:\n{feedback}\n\n[self-reflexion]:',
            )
        ]
        reflection = model.generate_chat(messages=messages)
        #print(f'Self reflection output: {reflection}')
    else:
        messages = [
            Message(
                role="system",
                content=instruction,
            ),
            Message(
                role="user",
                content=f"[Question]:\n{q}\n\n[Answer Generated]:\n{cur_answer}\n\n[Checklist feedback]:\n{feedback}\n\n[self-reflexion]:",
            )
        ]
        reflection = model.generate_chat(messages=messages)

    return reflection  # type: ignore
```
